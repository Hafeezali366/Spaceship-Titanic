```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV data into a pandas DataFrame
df = pd.read_csv('train.csv')
```

```python
# --- 1. Inspect the Shape ---
print(f"Dataset Shape: {df.shape}")
```

```python

# --- 2. Preview the Data ---
print("\nFirst 5 Rows:")
display(df.head())
```

```python
# --- 3. Check Data Types & Nulls ---
print("\nData Types and Missing Values:")
df.info()
```

```python
# Analyze the target variable
print(df['Transported'].value_counts(normalize=True))
sns.countplot(data=df, x='Transported')
plt.title('Distribution of Transported Passengers')
plt.show()
```

```python
# Descriptive statistics for numerical columns
print("Descriptive Statistics for Numerical Features:")
display(df.describe())

# Plot histograms for numerical features
numerical_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
df[numerical_cols].hist(figsize=(12, 8), bins=30)
plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col, hue='Transported', order=df[col].value_counts().index)
    plt.title(f'{col} vs. Transported')
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, x='Transported', y=col)
    plt.title(f'{col} vs. Transported')
    plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# --- Better Plot 1: Box Plot with Log Scale ---
plt.figure(figsize=(10, 6))
# We add a small constant (1) before taking the log to handle the zero values
df['Spa_log'] = np.log1p(df['Spa'])
sns.boxplot(data=df, x='Transported', y='Spa_log')
plt.title('Spa Spending (Log Scale) vs. Transported')
plt.ylabel('Spa (Log Scale)')
plt.show()
```

```python
# --- Better Plot 2: Violin Plot with Log Scale ---
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Transported', y='Spa_log') # Using the log-transformed data again
plt.title('Spa Spending Distribution (Log Scale) vs. Transported')
plt.ylabel('Spa (Log Scale)')
plt.show()
```

```python
# --- Better Plot 3: Bar Chart of Binary Spending ---
df['SpentOnSpa'] = df['Spa'] > 0

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='SpentOnSpa', hue='Transported')
plt.title('Transported Status by Whether Passenger Spent on Spa')
plt.xticks(ticks=[0, 1], labels=['Spent $0', 'Spent > $0'])
plt.show()
```
